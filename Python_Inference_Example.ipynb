```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

tokenizer = AutoTokenizer.from_pretrained("KIZervus/KIZervus")

model = AutoModelForSequenceClassification.from_pretrained("KIZervus/KIZervus", from_tf=True)
```

```python
pipe = pipeline("text-classification", model=model, tokenizer=tokenizer, framework="tf")
```

```python
preds = pipe("Ich liebe dich!")
```
